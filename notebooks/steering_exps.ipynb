```python
import pickle
import matplotlib.pyplot as plt
```

```python
# Load in steering result dictionary
with open('../src/steering_results.pkl', 'rb') as file:
    steering_results = pickle.load(file)
```

```python
# Sort results by layer
layer_18 = [steering_results['baseline'], steering_results['l18_0.25'], steering_results['l18_0.5'], steering_results['l18_0.75'], steering_results['l18_1.0']]
layer_22 = [steering_results['baseline'], steering_results['l22_0.25'], steering_results['l22_0.5'], steering_results['l22_0.75'], steering_results['l22_1.0']]
```

```python
steering_results['l18_0.75'] / steering_results['baseline']
```

```python
# Generate plot
x = [0, 0.25, 0.5, 0.75, 1.0]
plt.title("Steering Effect on Faithfulness at Different Layers")
plt.xlabel("Alpha Value")
plt.ylabel("Faithfulness Rate")
plt.plot(x, [100 * i for i in layer_18], label="Layer 18")
plt.plot(x, [100 * i for i in layer_22], label="Layer 22")
# plt.plot(x, [100 * i for i in layer_20], label="Layer 20")
# plt.plot(x, [100 * i for i in layer_24], label="Layer 24")
# plt.plot(x, [100 * i for i in layer_28], label="Layer 28")
plt.legend();
# plt.savefig("steering_results.png", dpi=300);
```
